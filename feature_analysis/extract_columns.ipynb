```python
import pandas as pd
```

```python
df = pd.read_csv('../source_data/extracted/articles.A-B.xml.tar.gz.csv')
df2 = pd.read_csv('../source_data/extracted/articles.C-H.xml.tar.gz.csv')
df3 = pd.read_csv('../source_data/extracted/articles.I-N.xml.tar.gz.csv')
df4 = pd.read_csv('../source_data/extracted/articles.O-Z.xml.tar.gz.csv')
```

```python
df['ack_absent'] = df.Acknowledgment_Tag.isna()

print(df)
```

```python
#Reduce the file complexity by dropping the Acknowledgment_Tag
df = df.drop(labels={'Acknowledgment_Tag'}, axis=1)

print(df)
```

```python
df.to_csv('../parsed_data/cleaned_data/articles.A-B.xml.tar.gz_reduced.csv')
```

```python
df2['ack_absent'] = df2.Acknowledgment_Tag.isna()
df2 = df2.drop(labels={'Acknowledgment_Tag'}, axis=1)
df2.to_csv('../parsed_data/cleaned_data/articles.C-H.xml.tar.gz_reduced.csv')

df3['ack_absent'] = df3.Acknowledgment_Tag.isna()
df3 = df3.drop(labels={'Acknowledgment_Tag'}, axis=1)
df3.to_csv('../parsed_data/cleaned_data/articles.I-N.xml.tar.gz_reduced.csv')

df4['ack_absent'] = df4.Acknowledgment_Tag.isna()
df4 = df4.drop(labels={'Acknowledgment_Tag'}, axis=1)
df4.to_csv('../parsed_data/cleaned_data/articles.O-Z.xml.tar.gz_reduced.csv')
```
